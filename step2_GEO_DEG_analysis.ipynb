```python
import pandas as pd

# Load the processed expression dataset
file_path = "/content/drive/MyDrive/Major Project/Prostate GEO  GSE45016.csv"  # Update with correct path if needed
data = pd.read_csv(file_path)

# Define relevant column names
logfc_col = "logFC"
pval_col = "P.Value"
adj_pval_col = "adj.P.Val"

# Define cutoff values
logfc_cutoff = 1
pval_cutoff = 0.05

# Identify upregulated and downregulated genes
upregulated = data[(data[logfc_col] >= logfc_cutoff) & (data[pval_col] < pval_cutoff)]
downregulated = data[(data[logfc_col] <= -logfc_cutoff) & (data[pval_col] < pval_cutoff)]

# Combine results
deg_data = pd.concat([upregulated, downregulated])
deg_data["Regulation"] = deg_data[logfc_col].apply(lambda x: "Upregulated" if x >= logfc_cutoff else "Downregulated")

# Save DEGs to local storage
deg_file = "/content/drive/MyDrive/Major Project/GSE45016_DEGs.csv"
deg_data.to_csv(deg_file, index=False)

print(f"Saved DEGs to {deg_file}")
print(f"Total DEGs: {len(deg_data)}")
print(f"Upregulated genes: {len(upregulated)}")
print(f"Downregulated genes: {len(downregulated)}")
```

```python
plt.figure(figsize=(4, 3))
sns.histplot(data[logfc_col], bins=50, kde=True, color="purple")
plt.axvline(x=logfc_cutoff, color="red", linestyle="dashed", linewidth=1, label="Upregulated Threshold")
plt.axvline(x=-logfc_cutoff, color="blue", linestyle="dashed", linewidth=1, label="Downregulated Threshold")

plt.xlabel("Log2 Fold Change (LogFC)")
plt.ylabel("Frequency")
plt.title("Distribution of LogFC Values")
plt.legend()
plt.show()
```
