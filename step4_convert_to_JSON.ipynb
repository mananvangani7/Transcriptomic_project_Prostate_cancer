```python
import pandas as pd
import json

# === Step 1: Load the merged CSV ===
csv_path = "/content/drive/MyDrive/Major Project/GEO_Papers_Merged_Fuzzy.csv"
df = pd.read_csv(csv_path)

# === Step 2: Prepare JSON records ===
records = []
for _, row in df.iterrows():
    record = {
        "Gene.symbol": row.get("Gene.symbol", ""),
        "Gene.title": row.get("Gene.title", ""),
        "Regulation": row.get("Regulation", ""),
        "Title": row.get("Title", ""),
        "Abstract": row.get("Abstract", ""),
        "PMID": row.get("PMID", "")
    }
    records.append(record)

# === Step 3: Save as JSONL ===
jsonl_path = "/content/drive/MyDrive/Major Project/finetune_data.jsonl"
with open(jsonl_path, "w", encoding="utf-8") as f:
    for record in records:
        json.dump(record, f)
        f.write('\n')

print(f"\n✅ JSONL file written successfully with {len(records)} records!")

# === Step 4: Verify JSONL by reading it back ===
print("\n🔍 Verifying the first 3 entries:")
with open(jsonl_path, "r", encoding="utf-8") as f:
    for i, line in enumerate(f):
        if i >= 3:
            break
        print(json.loads(line))
```
