```python
import json
import re

# === CHUNKING FUNCTION ===
def sliding_sentence_chunks(text, max_words=150, stride=75):
    sentences = re.split(r'(?<=[.!?]) +', text)
    words = [word for sentence in sentences for word in sentence.split()]
    chunks = []
    i = 0
    while i < len(words):
        chunk_words = words[i:i+max_words]
        chunk_text = " ".join(chunk_words)
        chunks.append(chunk_text)
        i += stride
    return chunks

# === FILE PATHS ===
input_file = "/content/drive/MyDrive/Major Project/finetune_data (step 4 result).jsonl"
output_file = "/content/drive/MyDrive/Major Project/chunked_data_v2.jsonl"

# === PROCESS FILE ===
chunk_count = 0
with open(input_file, "r", encoding="utf-8") as fin, \
     open(output_file, "w", encoding="utf-8") as fout:

    for line in fin:
        try:
            entry = json.loads(line.strip())
            gene = entry["Gene.symbol"]
            title = entry["Title"]
            abstract = entry["Abstract"]

            full_text = f"Gene: {gene}\nTitle: {title}\nAbstract: {abstract}"
            chunks = sliding_sentence_chunks(full_text)

            for i, chunk in enumerate(chunks):
                json.dump({"chunk": chunk, "doc_id": f"{gene}_{i}"}, fout)
                fout.write("\n")
                chunk_count += 1

        except KeyError:
            continue

# === CHECK OUTPUT ===
print(f"✅ Done! Total chunks written: {chunk_count}")

# Optional: Preview first few chunks
with open(output_file, "r", encoding="utf-8") as f:
    for _ in range(3):
        print(json.loads(f.readline()))
```
